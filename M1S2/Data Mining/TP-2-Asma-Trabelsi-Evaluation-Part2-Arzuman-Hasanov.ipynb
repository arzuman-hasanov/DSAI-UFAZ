# Clustering IRIS dataset using k-means

Now let's apply the same approach to iris dataset.

- Import datasets from sklearn; Load iris datset using .load_iris().

- Consider the predictors X the first two columns.

- Plot using scatter; Specify the c=target keyword arguments to color the points by their cluster label.

- Create k-means model; fit the model to X.

- Specify new labels; plot scatter using c=new_labels.

**Step 1:** Import packages

```python
from sklearn import datasets
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd
```

**Step 2:** load Iris dataset

```python
iris = datasets.load_iris()
```

**Step 3:** Consider the predictors X the first two columns and y should contain the iris is target class

```python
X = iris.data[:, :2]
y = iris.target
```

**Step 4:** Plot using scatter; Specify the c=target keyword arguments to color the points by their cluster label. 

```python
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='winter')
plt.xlabel('Length')
plt.ylabel('Width')
plt.title('Plot of Iris dataset')
plt.show()
```

**Step 5:** Create k-means model with k equals 3; fit the model to X.

```python
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)
```

**Step 6:** look for true_labels and new_labels

```python
true_labels = kmeans.labels_
new_labels = true_labels
```

**Step 7:** Plot scatter using c=new_labels

```python
plt.scatter(X[:, 0], X[:, 1], c=new_labels, cmap='winter')
plt.xlabel('Length')
plt.ylabel('Width')
plt.title('Plot after K-Means clustering')
plt.show()
```
