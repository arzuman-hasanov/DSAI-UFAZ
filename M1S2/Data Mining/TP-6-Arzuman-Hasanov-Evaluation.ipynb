# Data Mining: Clustering using kmeans

**Step 1:** import needed packages

```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import seaborn as sns
```

**Step 2:** Load the dataset (data_UFAZ/Wholesale customers data.csv). Output the first five rows to inspect the data content.

```python
# Load the dataset
data = pd.read_csv("Wholesale customers data.csv")

# Output the first five rows
print(data.head())
```

**Step 3:** print the statistics of the data

```python
# Print the statistics of the data
print(data.describe())
```

**Step 4:** standardizing the data

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

scaled_df = pd.DataFrame(scaled_data, columns=data.columns)
```

**Step 5:** Print the statistics of scaled data

```python
print(scaled_df.head())
```

**Step 6**: Define the kmeans function with 2 clusters and fit the k-means algorithm on scaled data

```python
def kmeans_clustering(data, n_clusters=2):

    kmeans = KMeans(n_clusters=n_clusters)

    kmeans.fit(data)
    
    return kmeans

num_clusters = 2

kmeans_model = kmeans_clustering(scaled_data, n_clusters=num_clusters)
```

**Step 7:** Use the inertia method to find the good number of clusters. Try clusters form 1 to 20

```python
inertia_values = []

for k in range(1, 21):

    kmeans = KMeans(n_clusters=k)
    kmeans.fit(scaled_data)
    
    inertia_values.append(kmeans.inertia_)


plt.plot(range(1, 21), inertia_values, marker='o')
plt.xlabel('numb of clusters')
plt.ylabel('inertia')
plt.title('inertia vs numb of clusters')
plt.show()
```

**Step 8:** Convert the results into a dataframe and plotting them

```python
results_df = pd.DataFrame({'Number of clusters': range(1, 21), 'Inertia': inertia_values})

plt.figure(figsize=(8, 6))
plt.plot(results_df['Number of clusters'], results_df['Inertia'], marker='o')
plt.xlabel('numb of clusters')
plt.ylabel('inertia')
plt.title('inertia vs numb of clusters')
plt.grid(True)
plt.show()
```

**Step 9**: Define the kmeans function with the best umber of clusters and fit the k-means algorithm on scaled data

```python
def kmeans_clustering_best(data, n_clusters):

    kmeans = KMeans(n_clusters=n_clusters)
    kmeans.fit(data)
    
    return kmeans

best_num_clusters = 5
kmeans_model_best = kmeans_clustering_best(scaled_data, n_clusters=best_num_clusters)
```

**Step 10 :** Predict the clusters of the scaled data

```python
predicted_clusters = kmeans_model_best.predict(scaled_data)
print(predicted_clusters)
```

**Step 11 :** Create a data frame that merge scaled data and their clusters 

```python
clustered_data = pd.concat([scaled_df, pd.DataFrame(predicted_clusters, columns=['Cluster'])], axis=1)
print(clustered_data.head())
```

**Step 12 :** Print the distribution of each cluster

```python
cluster_distribution = clustered_data['Cluster'].value_counts()
print(cluster_distribution)
```

# One Hot Encode

**Step 6:** Extract InvoiceNo and Description from the dataframe. Use then the groupby function and convert the Description field to a list

```python
data_new = data.groupby('InvoiceNo')['Description'].agg(list).reset_index()
print(data_new.head())
```

**Step 9** One hot encoding

```python
from mlxtend.preprocessing import TransactionEncoder

# Instantiate an encoder.
encoder = TransactionEncoder()

# Fit encoder to list of lists.
encoder.fit(data_new['Description'])

# Transform lists into one-hot encoded array.
onehot = encoder.transform(data_new['Description'])

# Convert array to pandas DataFrame.
onehot_df = pd.DataFrame(onehot, columns=encoder.columns_)
```

```python
print(onehot_df.head())
```

**Step 10** Compute the support metric

```python
# Print support metric over all rows for each column.
support = onehot_df.mean()
print(support)
```

**Step 11** Compute the item count distribution over transactions

```python
# Print distribution of item counts.
item_count_distribution = onehot_df.sum(axis=1).value_counts().sort_index()
print(item_count_distribution)
```

**Step 13** Compute joint support for COLOURED PARTY BALLOONS and MESSAGE CARDS WITH ENVELOPES.

```python
joint_support = (onehot_df['12 COLOURED PARTY BALLOONS'] & onehot_df['12 MESSAGE CARDS WITH ENVELOPES']).mean()

print(joint_support)
```

```python
antecedent_support = onehot_df['12 COLOURED PARTY BALLOONS'].mean()

confidence = joint_support / antecedent_support

print(confidence)
```

```python
# Calculate confidence for the rule "12 MESSAGE CARDS WITH ENVELOPES -> 12 COLOURED PARTY BALLOONS"
antecedent_support = onehot_df['12 MESSAGE CARDS WITH ENVELOPES'].mean()
joint_support = (onehot_df['12 MESSAGE CARDS WITH ENVELOPES'] & onehot_df['12 COLOURED PARTY BALLOONS']).mean()

confidence = joint_support / antecedent_support

# Display the confidence
print(confidence)
```

**Step 17** Use apriori() to identify frequent itemsets Apply apriori algorithm to data with min support threshold of 0.03

```python
from mlxtend.frequent_patterns import apriori

frequent_itemsets = apriori(onehot_df, min_support=0.03, use_colnames=True)

# Print frequent itemsets
print(frequent_itemsets)
```

**Step 18** Use apriori() to identify frequent itemsets Apply apriori algorithm to data with min support threshold of 0.02

```python
frequent_itemsets = apriori(onehot_df, min_support=0.02, use_colnames=True)

# Print frequent itemsets
print(frequent_itemsets)
```

**Step 19** Use apriori() to identify frequent itemsets Apply apriori algorithm to data with min support threshold of 0.01

```python
frequent_itemsets = apriori(onehot_df, min_support=0.01, use_colnames=True)

# Print frequent itemsets
frequent_itemsets
```

**Step 20** Use association_rules() to compute and prune association rules from output of apriori().
Recover association rules using support and a minimum threshold of 0.01.

```python
from mlxtend.frequent_patterns import association_rules
rules = association_rules(frequent_itemsets, metric='support', min_threshold=0.01)

# Print rules header
rules.head()
```

**step 21** Recover association rules using confidence and a minimum threshold of 0.8.

```python
rules = association_rules(frequent_itemsets, metric='confidence', min_threshold=0.8)

# Print rules header
rules.head()
```

**Step 22** Select rules with a consequent support above 0.08.

```python
rules_selected = rules[rules['consequent support'] > 0.08]
rules_selected
```

**Step 23** Select rules with leverage higher than 0.01.

```python
rules_selected = rules[rules['leverage'] > 0.01]

# Print rules
print(rules_selected)
```
